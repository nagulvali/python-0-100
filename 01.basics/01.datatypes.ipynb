# Data types
- A data type in programming language means type of value of the data
- python has below built-in data types
    - Numeric Types: `int`, `float`, `complex`
    - Text Types: `str`
    - Sequence Types: `list`, `tuple`, `range`
    - Mapping Types: `dict`
    - Set Types: `set`, `frozenset`
    - Boolean Types: `bool`
    - Binary Types: `bytes`, `bytearray`, `memoryview`
    - None Type: `NoneType`
- You can get the data type of any object by using the `type()` function
-


```python
# Data Types

## numeric types
a = 10
b = 20.50
c = 1 + 2j

print(type(a))
print(type(b))
print(type(c))

## Text Types
txt = "Hello World !!"
print(type(txt))

## Sequence Types

## list:
# - ordered with index
# - changeable
# - Allows duplicate values
ls = [1, 2, 3, 4, 5]
print(ls)
print(type(ls))

## tuple:
# - ordered with index
# - un-changeable
# - allows duplicate values
thistuple = ("c", "d", "b", "a")
print(thistuple)
print(type(thistuple))

## Dictionaries:
# - used to store key, value pairs
# - python 3.7+ dictionaries are ordered, 3.6 and below version python dicts are not ordered
# - changeable
# - do not allows duplicate keys
#
thisdict = {
    "a": "hellow",
    "a": "hello world"
}
print(thisdict)

## Set
# - unordered
# - un-changeable
#   - you can add, delete items
# - un-indexed
# - do not allow duplicate values

thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)

## Frozen Set:
# - unordered (no-index)
# - un-changeable: you can not even add, delete items once created (read-only)
# - do not allow duplicate values

mylist = ['apple', 'banana', 'cherry']
x = frozenset(mylist)
# x[1] = "strawberry"


## bin() function
# bin function converts binary version of the data type
```
